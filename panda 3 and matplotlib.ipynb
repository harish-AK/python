```python
import pandas as pd
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                'B':['B0','B1','B2','B3'],
                'C':['C0','C1','C2','C3'],
                'D':['D0','D1','D2','D3']},
                index=[0,1,2,3])
```

```python
df1
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                'B':['B4','B5','B6','B7'],
                'C':['C4','C5','C6','C7'],
                'D':['D4','D5','D6','D7']},
                index=[4,5,6,7])
```

```python
df2
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                'B':['B8','B9','B10','B11'],
                'C':['C8','C9','C10','D11'],
                'D':['D8','D9','D10','D11']},
                index=[8,9,10,11])
```

```python
df3
```

```python
#concatenation
pd.concat([df1,df2,df3],axis=0) #add row wise is default
```

```python
pd.concat([df1,df2,df3],axis=1)  # add column wise
```

```python
#example data frames
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                  })
right=pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'C':['C0','C1','C2','D3'],
                    'D':['D0','D1','D2','D3']})
                   
                    
```

```python
left
```

```python
right
```

```python
#merging
pd.merge(left,right,on='key')    # should have a same column in both datas to merge
```

```python
left=pd.DataFrame({'key1':['k0','k0','k1','k2'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                  })
right=pd.DataFrame({'key1':['k0','k0','k1','k2'],
                    'key2':['K0','K1','K0','K1'],
                    'C':['C0','C1','C2','D3'],
                    'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

```python
pd.merge(left,right,how='right',on=['key1','key2'])
```

```python
#joining
left=pd.DataFrame({'A':['A0','A1','A2'],
                'B':['B0','B1','B2']},index=['k0','k1','k2'])
right=pd.DataFrame({'C':['C0','C2','C3'],
                'D':['D0','D2','D3']},
                index=['k0','k2','k3'])
```

```python
left
```

```python
right
```

```python
left.join(right,how='inner')
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```

```python
left.join(right) # non matching values on right is nan
```

```python
left.join(right,how='left')
```

```python
left.join(right,how='right')
```
