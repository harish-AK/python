```python
data=[10,5,6,1,3,7,9,4,8,2]
mean=sum(data)/len(data)
mean
```

```python
lst=[]
for i in data:
    var=i-mean
    lst.append(var)
print(lst)
```

```python
sq=[]
for j in lst:
    sq.append(j*j)
print(sq)
```

```python
a=sum(sq)
b=len(data)-1
variance=a/b
variance
```

```python
import math
sd=math.sqrt(variance)
sd
```

```python
zscore=[]
for k in data:
    z=k-mean
    z=z/sd
    zscore.append(z)
print(zscore)
```
