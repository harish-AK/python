```python
#legends
fig=plt.figure()
x=np.arange(-10,10)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square',color='red')
ax.plot(x,x**3,label='cube',color='green')
ax.plot(x,np.sqrt(x),label="sq root",color='yellow')
ax.legend()
```

```python
ax.legend(loc=1) # up right corner
ax.legend(loc=2) # up left corner
ax.legend(loc=3) # low right corner
ax.legend(loc=4) # low left corner
ax.legend(loc=0)
fig
```
