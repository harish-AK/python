```python
#multi index and index hierarchy
outside=['g1','g1','g1','g2','g2','g2','g3']
inside=[1,2,3,1,2,3,1]
hier=list(zip(outside,inside))
hier=pd.MultiIndex.from_tuples(hier)
```

```python
hier
```

```python
df=pd.DataFrame(np.random.randn(7,2),index=hier,columns=['A','B'])
```

```python
df
```

```python
df.loc['g1']
```

```python
df.index.names=['group','num']
df
```
